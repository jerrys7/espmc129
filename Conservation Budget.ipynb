```python
h = [0.1, 0.3, 0.5]
```

```python
# initial conditions
c0 = 350e-6
flux = 2e-6
air_density = 101275 / (298 * (8.31))
time_steps = 1000
g = 4.3 * 10e-4
```

```python
air_density
```

```python
height_carbon = []
height_flux = []
```

```python
for height in h:
    carbon_concentration = [c0]
    flux_densities = [flux]
    for i in range(time_steps):
        dc = flux_densities[i] / (height * air_density)
    #     changes.append(next_change)
        next_ct = carbon_concentration[i] + dc
        carbon_concentration.append(next_ct)
        next_flux = -g * (next_ct - 5000e-6)
        flux_densities.append(next_flux)
    height_carbon.append(carbon_concentration)
    height_flux.append(flux_densities)
```

```python
time = [i for i in range(1000)]
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(len(h)):
    plt.plot(time, height_flux[i][1:])
```

```python
for i in range(len(h)):
    plt.plot(time, height_carbon[i][1:])
```
